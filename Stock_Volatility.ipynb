```python
import yfinance as yf

from yahoo_fin import stock_info as si

nasdaq_tickers = si.tickers_sp500()


tickers = nasdaq_tickers # Replace with the tickers of the companies you are interested in

dfs = {}
for ticker in tickers:
    dfs[ticker] = yf.download(ticker, start="2023-04-04", end="2023-04-05")
```

```python
returns = {}
for ticker in tickers:
    returns[ticker] = dfs[ticker]["Adj Close"].pct_change()
    
print(returns)
```

```python
std_devs = {}
for ticker in tickers:
    std_devs[ticker] = returns[ticker].std()
```

```python
sorted_std_devs = sorted(std_devs.items(), key=lambda x: x[1], reverse=True)
```

```python
for ticker, std_dev in sorted_std_devs:
    print(f"{ticker}: {std_dev}")
```
